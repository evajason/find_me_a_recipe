```python
ingredients_lists = raw_recipes_df['ingredients'].apply(lambda x: ast.literal_eval(x))
ingredients_df = ingredients_lists.explode().value_counts().to_frame().reset_index().rename(columns={'ingredients':'ingredient'})
```

```python
recipes_df = raw_recipes_df.copy()
recipes_df['tags_list'] = raw_recipes_df['tags'].apply(ast.literal_eval)
```

```python
recipes_df['n_tags'] = recipes_df['tags_list'].apply(len)
```

```python
recipes_df.head()
```

```python
# average number of tags
recipes_df['n_tags'].mean()
```

##### Tags

```python
tags = recipes_df[['tags_list','id']].explode('tags_list').rename(columns={'tags_list':'tag'})
tags
```

```python
# number of tags only mentioned once
tag_counts = tags.groupby('tag').count()#.shape[0]
tag_counts[tag_counts['id'] == 1].shape[0]
```

```python
# average rating for each itemid
ratings_per_recipe = raw_interactions_df[['recipe_id', 'rating']].groupby('recipe_id').mean()
ratings_per_recipe.head()
```

```python
tag_ratings = tags.merge(ratings_per_recipe, left_on='id', right_index=True)
tag_ratings = tag_ratings[['tag', 'rating']].groupby('tag').mean().sort_values(by='rating', ascending=False).reset_index()
```

```python
tag_ratings
```

```python
# filter off tags that are used less than 5 times
rare_tags = set(tag_counts[tag_counts['id'] < 5].index)
tag_ratings[tag_ratings['tag'].apply(lambda x: x not in rare_tags)]
```
